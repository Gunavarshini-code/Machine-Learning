```python
#data understanding/loading
import pandas as pd
```

```python
d=pd.read_csv("salary_data.csv")
d
```

```python
d.describe()
```

```python
#feature selection/engineering(here we tell feature data and target data)
x=d[['YearsExperience']]
x
y=d.iloc[:,1]
x
```

```python
#Train Test Spliting (split data before training a model)
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
x_train
```

```python
x_test
```

```python
#Model Selection / Model Training stage
from sklearn.linear_model import LinearRegression
```

```python
lr=LinearRegression()
lr.fit(x_train,y_train)#Trains the model using training data
```

```python
#Model Evaluation / Prediction stage
y_pred=lr.predict(x_test)
y_pred
```

```python
y_test
```

```python
#Model Evaluation stage
from sklearn.metrics import r2_score
```

```python
r2_score(y_test,y_pred)
```

```python
years = float(input("Enter Years of Experience: "))
years = [[years]]  # reshape to 2D
salary = lr.predict(years)
print("Predicted Employee Salary:", salary[0])
```
