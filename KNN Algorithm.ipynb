```python
#Data Understanding/Loading
import pandas as pd
```

```python
d=pd.read_csv("iphone_purchase_records.csv")
d
```

```python
#Data Cleaning/Preprossing
d.isnull().sum()
```

```python
d.describe()
```

```python
d.info()
```

```python
#Label Encoding(Preprocessing technique)
from sklearn.preprocessing import LabelEncoder
```

```python
l=LabelEncoder()#instantiation(or) object creation
d['Gender']=l.fit_transform(d['Gender'])
d['Gender']
```

```python
d
```

```python
#feature selection/engineering(here we tell feature data and target data)
x=d[["Gender","Age","Salary"]]
y=d["Purchase Iphone"]
```

```python
#model selection (split data before training a model)
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
x_train
```

```python
x_test
```

```python
#Feature Scaling using StandardScalar
from sklearn.preprocessing import StandardScaler
ss=StandardScaler()
ss_train=ss.fit_transform(x_train)
ss_train
```

```python
ss_test=ss.transform(x_test)
ss_test
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
kn=KNeighborsClassifier(n_neighbors=5,metric="euclidean")
kn.fit(ss_train,y_train)
```

```python
y_pred=kn.predict(ss_test)
y_pred
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
```

```python
accuracy_score(y_pred,y_test)
```

```python
confusion_matrix(y_pred,y_test)
```

```python
classification_report(y_pred,y_test)
```
